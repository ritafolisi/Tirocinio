# Setosa

```python
df = pd.read_csv('iris-setosa.csv')

df = df.sample(frac=1)

X = df.iloc[:, 1:3].values
y = df.iloc[:,0].values
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(X,y)
model = FuzzyKNN()
```

```python
a = np.arange (1, 21, 2)
parameters = {"k" : a}
parameters["k"]
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(model, parameters, cv = 5)
```

```python
clf.fit(xTrain, yTrain)
```

```python
model = clf.best_estimator_
```

```python
clf.score(xTest, yTest) 
```

```python
from sklearn.preprocessing import StandardScaler
iris_values_std = StandardScaler().fit_transform(X)

from sklearn.decomposition import PCA
pca_2d = PCA(n_components=2)
iris_values_2d = pca_2d.fit_transform(iris_values_std)
```

```python
#stampa
%matplotlib inline
import matplotlib.pyplot as plt


def gr_dataset(): 
    for lab, col in zip((1, 0),
                        ('blue', 'red')):
        plt.scatter(iris_values_2d[y==lab, 0],
                    iris_values_2d[y==lab, 1],
                    label=lab,
                    c=col)

gr_dataset()
```

```python
def gr_membership_contour(estimated_membership):
    p = np.arange(-4, 4, .1)
    l = np.arange(-4, 4, .1)
    P, L = np.meshgrid(p, l)
    res = []
    res_memb = []
    for p,l in zip(np.ravel(P), np.ravel(L)): 
        memb, _ = estimated_membership((p, l))
        res_memb.append(memb[0])
    for t in res_memb:
        res.append(t[1])
    zs = np.array(res)
    print(f'Contenuto di zs = {np.unique(zs)}')
    Z = zs.reshape(P.shape)
    membership_contour = plt.contour(P, L, Z,
                                     levels=(.5, .7, .9), colors='k')
    plt.clabel(membership_contour, inline=1)
```

```python
def generator(m):
    return (-4 + np.random.random(2*m) * 8).reshape((m, 2))
```

```python
plt.figure(figsize=(7,7))
gr_dataset()
gr_membership_contour(model.predict)
plt.show()
```
